```python
print(" The actor generating the highest average revenue?")
df_q2=df_csv[['Actors','Revenue (Millions)']] 
df_q2=df_q2.fillna('Nope') 
Act=df_q2['Actors'].str.split("|") 
Act_list=[] 
for i in range(len(Act)): 
    for j in range(len(Act[i])): 
        Act_list.append(Act[i][j].strip())

Act_uq=pd.unique(Act_list).tolist() 
Act_uq 
Act_rev_avg=[] 


for i in range(len(Act_uq)): 
    temp=[] 
    for j in range(len(df_q2)): 
        if Act_uq[i] in df_q2['Actors'][j] and df_q2['Revenue (Millions)'][j]!='Nope': 
            temp.append(df_q2['Revenue (Millions)'][j]) 
    Act_rev_avg.append(np.mean(temp))

df_act_rev=pd.DataFrame(Act_uq,columns=['Actors'])
df_act_rev=pd.concat([df_act_rev,pd.DataFrame(Act_rev_avg,columns=['Revenue (Millions)'])],axis=1)
df_act_rev=df_act_rev.fillna(0)
df_act_rev_sort=df_act_rev.sort_values(by=['Revenue (Millions)'],ascending=False)
df_act_rev_sort
max_num=np.max(df_act_rev_sort['Revenue (Millions)'])
max_num
count=0
for i in range(len(df_act_rev_sort)):
    if df_act_rev_sort['Revenue (Millions)'][i]==max_num:
        count+=1
        print(df_act_rev_sort['Actors'][i],df_act_rev_sort['Revenue (Millions)'][i])
print("Total=",count)
```

```python
print("Top‐3 directors who collaborate with the most actors?")
df_q4=df_csv[['Director','Actors']]
df_act=df_q4['Actors'].str.split("|",expand=True)
df_act=df_act.fillna('Nope')
df_dir=df_q4['Director']
df_dir=pd.DataFrame(df_dir,columns=['Director'])
df_dir_act=pd.concat([df_dir,pd.DataFrame(df_act)],axis=1)
df_dir_act=df_dir_act.sort_values(by=['Director'],ascending=True)
gp_dir=df_dir_act.groupby("Director")
dir_list=df_q4['Director'].drop_duplicates().tolist()
dir_list=sorted(dir_list)
dir_list
act_list=df_q4['Actors'].tolist()
act_list
for i in range(len(act_list)):
    act_list[i]=act_list[i].replace("| ","|").split("|")
act_list
dir_co=[]
for i in range(len(dir_list)):
    co_temp=[]
    for j in range(len(df_dir_act)):
        if dir_list[i]==df_dir_act['Director'][j]:
            for k in range(len(act_list[j])):
                 co_temp.append(act_list[j][k])
    co_temp=pd.unique(co_temp).tolist()
    dir_co.append(len(co_temp))
df_dir_co=pd.DataFrame(dir_list,columns=['Director'])
df_dir_co=pd.concat([df_dir_co,pd.DataFrame(dir_co,columns=['Co_num'])],axis=1)
df_dir_co
df_dir_co=df_dir_co.sort_values(by=['Co_num'],ascending=False)
df_dir_co
max=df_dir_co.nlargest(3,'Co_num')['Co_num'].tolist()
count=0
for i in range(len(max)):
    for j in range(len(df_dir_co)):
        if max[i]==df_dir_co['Co_num'][j]:
            count+=1
            print(df_dir_co['Director'][j],':',df_dir_co['Co_num'][j])
print("Total=",count)
```

```python
print(' Top‐2 actors playing in the most genres of movies?')
df_q5=df_csv[['Actors','Genre']]  
Act=df_q5['Actors'].str.split("|") 
Act_list=[] 
for i in range(len(Act)): 
    for j in range(len(Act[i])): 
        Act_list.append(Act[i][j].strip())

Act_uq=pd.unique(Act_list).tolist() 
Act_uq 
gen=df_q5['Genre'].tolist()
for i in range(len(gen)):
    gen[i]=gen[i].split("|")
gen
play_gen=[]
for i in range(len(Act_uq)):
    temp=[]
    for j in range(len(df_q5)):
        if Act_uq[i] in df_q5['Actors'][j]:
            for k in range(len(gen[j])):
                temp.append(gen[j][k])
    temp=pd.unique(temp).tolist()
    play_gen.append(len(temp))
play_gen
df_act_gen=pd.DataFrame(Act_uq,columns=['Actor'])
df_act_gen=pd.concat([df_act_gen,pd.DataFrame(play_gen,columns=['Genre'])],axis=1)
df_act_gen=df_act_gen.sort_values(by=['Genre'],ascending=False)
df_act_gen
max=df_act_gen.nlargest(2,'Genre')['Genre'].tolist()
max
count=0
for i in range(len(max)):
    for j in range(len(df_act_gen)):
        if max[i]==df_act_gen['Genre'][j]:
            count+=1
            print(df_act_gen['Actor'][j],":",df_act_gen['Genre'][j])
print("Total=",count)
```

```python
print('Top‐3 actors whose movies lead to the largest maximum gap of years?')
df_q6=df_csv[['Actors','Year']]
Act=df_q6['Actors'].str.split("|") 
Act_list=[] 
for i in range(len(Act)): 
    for j in range(len(Act[i])): 
        Act_list.append(Act[i][j].strip())

Act_uq=pd.unique(Act_list).tolist() 
Act_uq
year_gap=[]
for i in range(len(Act_uq)):
    temp=[]
    for j in range(len(df_q6)):
        if Act_uq[i] in df_q6['Actors'][j]:
            temp.append(df_q6['Year'][j])
    year_gap.append(np.max(temp)-np.min(temp))
df_act_yeargap=pd.DataFrame(Act_uq,columns=['Actor'])
df_act_yeargap=pd.concat([df_act_yeargap,pd.DataFrame(year_gap,columns=['MaxYearGap'])],axis=1)
df_act_yeargap=df_act_yeargap.sort_values(by=['MaxYearGap'],ascending=False)
max=df_act_yeargap.nlargest(3,'MaxYearGap')['MaxYearGap'].tolist()
max=pd.unique(max).tolist()
count=0
for i in range(len(max)):
    for j in range(len(df_act_yeargap)):
        if max[i]==df_act_yeargap["MaxYearGap"][j]:
            count+=1
            print(df_act_yeargap['Actor'][j],":",df_act_yeargap['MaxYearGap'][j])


print("Total=",count)
```

```python
print("Find all actors who collaborate with Johnny Depp in direct and indirect ways ")
df_q7=df_csv['Actors']

act_list=df_q7.tolist()
for i in range(len(act_list)):
    act_list[i]=act_list[i].split("|")
    for j in range(len(act_list[i])):
        act_list[i][j]=act_list[i][j].strip()
dirco=[]
for i in range(len(act_list)):
    if 'Johnny Depp' in act_list[i]:
        for j in range(len(act_list[i])):
            dirco.append(act_list[i][j])
dirco=pd.unique(dirco).tolist()
del dirco[0]
dirco
df_dirco=pd.DataFrame(dirco,columns=['direct_collaborate'])

gap=1
while gap>0:
    start=len(dirco)
    for i in range(start):
        for j in range(len(act_list)):
            if dirco[i] in act_list[j]:
                for k in range(len(act_list[j])):
                    dirco.append(act_list[j][k])
                   
    dirco=pd.unique(dirco).tolist()
    
    end=len(dirco)
    gap=end-start
df_indirco=pd.DataFrame(dirco,columns=['indirect_collaborate(含direct_collaborate)'])
df_ans=pd.concat([df_dirco,df_indirco],axis=1)
df_ans




            
```
