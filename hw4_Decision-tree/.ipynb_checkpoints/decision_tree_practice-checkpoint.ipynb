```python
#df_shroom=df_shroom[df_shroom.odor=="n"]

def information_gain(df, split_feature_name, target_feature_name):
    
    ### take a dataFrame of features, and quantify the entropy of a target feature
    ### after performing a split along the values of another feature
    ### and calculate the corresponding information gain
    

    # your code here
    
    info_gain=0
    prob_joint=0
    
    #先對split_feature_name的特徵進行數量排序
    sort=df[split_feature_name].value_counts()
    
    for i in range(len(sort)):
        
        #先找出split_feature_name 中有那些特徵,並找出其數量
        temp=sort.keys()[i]
        weights=sort[i]/len(df)
        #print(temp)        
        #print(weights)
        
        #紀錄取交集後的機率        
        num_temp=df[df[split_feature_name]==temp][target_feature_name].value_counts(1)
        #print(num_temp)
    
        dict_add={}    
        for i in range(len(num_temp)):
            name=num_temp.keys()[i]+","+temp
            dict_add[name]=num_temp[i]
        
        #print(dict_add)
        
        #利用加權算出entrophy
        prob_add=0
        if len(dict_add)>1:
            sum=0
            for i in range(len(dict_add)):
                add=list(dict_add.values())[i]
                #print(add)
                prob_add=weights*(-add*math.log(add,2))
                sum+=prob_add
                
             
            #print(sum)
            prob_joint+=sum
    
    #print(prob_joint)
    info_gain=total_entrophy-prob_joint
        
        
        
  #========================================================================================================      
        #針對split_feature_name中的特徵，找其對應的target_feature
        #num_poi=df[(df[split_feature_name]==temp) & (df[target_feature_name]=="poisonous")].shape[0]
        #num_edi=df[(df[split_feature_name]==temp) & (df[target_feature_name]=="edible")].shape[0]
        
        #temp_prob=df[split_feature_name].value_counts()[temp]/df.shape[0]
        #prob_poi=num_poi/(num_poi+num_edi)
        #prob_edi=num_edi/(num_poi+num_edi)
               
        #info_split_feature=0
        #if prob_poi!=0 and prob_edi!=0:
            #info_split_feature=temp_prob*(-prob_poi*math.log(prob_poi,2)-prob_edi*math.log(prob_edi,2))
        #else:
            #info_split_feature=0
        
        #info_gain+=info_split_feature
            
    return info_gain

#print(information_gain(df_shroom, 'odor', 'class'))
print('\nExample: information gain for the best feature is ' + str(information_gain(df_shroom, 'odor', 'class')))
```

```python
def id3(df, target_feature_name, feature_names, default_class = None):
    
    ## counting for the target feature
    from collections import Counter
    cnt = Counter(x for x in df[target_feature_name])
    
    ## First check: do instances in this split of the dataset belong to the same class?
    # (i.e., all mushrooms in this set are poisonous)
    # if yes, return that homogenous label (e.g., 'poisonous')
    if len(cnt) == 1:
        return list(cnt.keys())[0]
    
    ## Second check: is this split of the dataset empty?
    # if yes, return a default value
    elif df.empty or (not feature_names):
        return default_class 
    
    ## Otherwise: this dataset is ready to be split!
    else:
        ### step 1: get the default value for next recursive call of this function
        index_of_max = list(cnt.values()).index(max(cnt.values())) 
        default_class = list(cnt.keys())[index_of_max] # most common value of target feature in dataset (poisonous or edible)
        
        ### step 2: choose the best feature to split on
        
        # your code here
        #找出ig值最大的作為best_feature
        info_gain_feature=[]
        for i in range(len(feature_names)):
            info_gain_feature.append(information_gain(df,feature_names[i], 'class'))
        #print(info_gain_feature)
        max_ig=max(info_gain_feature)
        print(max_ig)
        max_index=info_gain_feature.index(max_ig)
        best_feature=feature_names[max_index]
        print(best_feature)
        
        #選定best_feature的每個特徵
        split=list(df[best_feature].unique())
        split.sort()

        #做出split 以及 split_value
        split_value=[]
        for i in range(len(split)):
            judge=df[best_feature]==split[i]
            poi_or_edi=list(set(list(df[judge][target_feature_name])))
            #print(poi_or_edi)
            if len(poi_or_edi)==1:
                split_value.append(poi_or_edi[0])
            else:
                split_value.append("more_split")
        
        print(split)
        print(split_value)
        

        
        ### step 3: create an empty tree, to be populated in a moment
        
        # your code here
        
        tree={}
        subtree = {split[i]:split_value[i] for i in range(len(split))}
        print(subtree)
        
        def get_key (dict, value):
            return [k for k, v in dict.items() if v == value]
        get_list=get_key(subtree,'more_split')
        print(get_list)
        
        
        
        #print(subtree)
        ### Step 4: split dataset
        # on each split, recursively call this "id3" function
        # populate the empty tree with subtrees, which
        # are the result of the recursive call
        
        #找出哪個特徵需要是more_split,需要生子樹根
        feature_names.remove(best_feature)
        df=df.drop(columns=[best_feature])
        
        
        
        for i in range(len(get_list)):
            subtree[get_list[i]]=subtree
        
 
    
        # your code here

         
            
                   
        return tree
```
